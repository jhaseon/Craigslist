```
import requests
import re 
```

```
page = requests.get('https://newyork.craigslist.org/search/bka')
```

```
numbers = [120, 240, 360, 480, 600, 720, 840, 960,1080]
URLS=['https://newyork.craigslist.org/search/bka']
for number in numbers:
  URLS.append('https://newyork.craigslist.org/search/bka' + "?s=" + str(number))
```

```
pages = []
for URL in URLS:
  pages.append(requests.get(URL))
```

```
list_dicts = []
titles =[]
prices = []

for page in pages:
  titles.append(re.findall(r'class="result-title hdrlnk">(.*?)<',str(page.content)))
  prices.append(re.findall(r'class="result-price">(.*?)<',str(page.content)))
# for dicts in list_dicts:
```

```
flat_list_pages = [item for sublist in titles for item in sublist][1:1001]
flat_list_pages2 = [item for sublist in prices for item in sublist][1:1001]
```

```
case_list = {}
for page in pages:
    case = {'title': flat_list_pages, 'price': flat_list_pages2}
    case_list.update(case)
```

```
import pandas as pd 
craigslist_df = pd.DataFrame.from_dict(case_list)
```
